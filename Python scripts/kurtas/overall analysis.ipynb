```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\devan\Desktop\CODING\Diwali_Clothes_Dataset.csv')
print(df.head())
```

```python
print('Plot for total sales, product wise')
total_sales = df.groupby(['Product_Name'], as_index = False)['Quantity_Sold'].count()
print(total_sales)
ax = sns.barplot(total_sales, x = 'Product_Name', y = 'Quantity_Sold', color = "blue")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
print('Total amount spent per product category')
total_amount = df.groupby(['Product_Name'], as_index = False)['Price'].sum().sort_values(by='Price', ascending = False)
print(total_amount)
ax = sns.barplot(data = total_amount, x = 'Product_Name', y = 'Price', color = 'green')
for container in ax.containers:
    ax.bar_label(container)
plt.title('Prices of products category wise')
plt.xlabel('Product Name')
plt.ylabel('Price')
plt.show()
```

```python
print('male and female distribution ka poe chart')
gender_counts = df.groupby('Gender')['CustomerID'].count()
print(gender_counts)
ax = gender_counts.plot.pie(autopct = '%1.2f%%')
plt.show()
```

```python
print('lineplot for Sales trend over time ')
ax = sns.lineplot(data = df, x='Date', y='Price',color = 'black', linestyle = 'dashed', marker = 'o')
plt.xticks(rotation=40)
plt.show()
```

```python
print('Price distribution')
price_dist = df.groupby(['Product_Name'], as_index = False)['Price'].sum()
print(price_dist) 
```

```python
print('top 5 states with most sales of clothes')
top_states = df.groupby(['State'], as_index = False)['Price'].sum().sort_values(by = 'Price', ascending = False)
print(top_states)
colors = ['cyan', 'brown', 'orange', 'coral', 'lime']
plt.figure(figsize = (7,5))
ax = sns.barplot(data = top_states, x = 'State', y = 'Price', palette=colors)
for container in ax.containers:
    ax.bar_label(container)
plt.title('Top states with highest sales')    
plt.show()

```
