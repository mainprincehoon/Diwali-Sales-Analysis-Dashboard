```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\devan\Desktop\CODING\Diwali_Sweets_Dataset.csv')
print(df.head())
```

```python
print('poe chart for count of orders category wise')
orders = df.groupby(['Product_Name'], as_index = True)['Quantity_Sold'].count()
print(orders)

orders.plot(kind = 'pie', y = 'Quantity_Sold', autopct = '%1.2f%%', colors = ['red', 'green', 'blue', 'yellow', 'orange'])

plt.title('Distribution of orders per category')
plt.ylabel(' ')
plt.legend(title = 'Product_Name', loc = 'center right', bbox_to_anchor = (1.5,0.5))
plt.tight_layout()
plt.show()
```

```python
print('sweets preferences gender wise: ')
male = df[df['Gender']=='Male'].groupby('Product_Name')['Gender'].count()
female = df[df['Gender']=='Female'].groupby('Product_Name')['Gender'].count()

print('For Males:','\n',male,'\n','For females:','\n',female)

combine_gender = pd.DataFrame({'Male': male, 'Female': female})
ax = combine_gender.plot(kind = 'bar', figsize = (10,6), color = ['blue', 'pink'])

plt.title('Sweet preferences by gender')
plt.xlabel('Category of sweets')
plt.ylabel('Gender')
plt.xticks(rotation = 0)
plt.legend(title = 'Gender')
plt.show()


```

```python
print('Day wise trend of sales')

daily_prices = df.groupby('Date')['Price'].sum()
print(daily_prices)

plt.figure(figsize = (12,6))
ax = sns.lineplot(data = daily_prices, linestyle = 'dotted', color = 'black', marker = 'o')

plt.title('Day wise trend of sales')
plt.xlabel('Date')
plt.ylabel('Amount spent')    
plt.tight_layout()
plt.plot()
```

```python
print('state wise sales of sweets:')

state_wise_sales = df.groupby(['State'], as_index = True)['Price'].sum().sort_values(ascending = True).head(5)
print(state_wise_sales)

ax = state_wise_sales.plot(kind = 'barh', figsize = (10,5), color = 'purple')
# ax = plt.barh(data = state_wise_sales, x = 'Price', y = 'State' ,color = 'purple')
for container in ax.containers:
    ax.bar_label(container)

plt.title('State wise sales of sweets')
plt.xlabel('Amount spent')
plt.ylabel('State')
plt.show()
```

```python
print('Profit and Revenue earned per category')
total_revenue = df.groupby(['Product_Name'], as_index = True)[['Total_Amount', 'Profit']].sum().round(2)
print(total_revenue)
ax = total_revenue.plot(kind = 'bar', stacked = True, color = ['black', 'grey'], figsize = (10,5))
plt.title('Revenue and profit')
plt.xlabel('Product Category')
plt.ylabel('price')
plt.xticks(rotation = 0)
plt.legend(loc = 'upper left', bbox_to_anchor = (1,0.5))
plt.tight_layout()
plt.show()

```
